## Solar calculations

Solar irradiance (source: https://ilmajaam.physic.ut.ee/)\
2023 Aprill til August

mean irradiance at 8-20 approx. 450 W/m2

Average predicted panel output 2.52 W

```python
import csv
import numpy as np
from scipy.integrate import trapezoid
from datetime import datetime, time


def read_column(csv_file):
    start_time = time(8,0,0)
    end_time = time(20,0,0)
    first_column = []
    second_column = []
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            # Assuming at least 2 columns in each row
            if len(row) >= 2:
                date = datetime.strptime(row[0], '%Y-%m-%d %H:%M:%S')
                if date.time() > start_time and date.time() < end_time:
                    second_column.append(row[1])
    return first_column, second_column

# Example usage:
csv_file = 'irradiance2023-aprill-august.csv'  # Provide your CSV file name here

first_column, second_column = read_column(csv_file)
print("First column:", first_column[1:])
print("Second column:", second_column[1:])

second_column = second_column[1:]
clean_values = []
for i in second_column:
    try:
        clean_values.append(float(i))
    except ValueError:
        pass


print(np.mean(clean_values))

x = np.linspace(0, 1, len(clean_values))

integral =  trapezoid(clean_values, x)
print(integral)
```
